## STEP 3 ‚Äî Data Cleaning & Preparation

This is the phase of our engineering pipeline which cleanses our dataset and makes it ready for future analysis.
We check missing values, rectify or drop them, and also deal with duplicates and outliers.

**Bad data ‚Üí misleading charts ‚Üí wrong business decisions.**

```python
# check missing values
print(df.isnull().sum())

# remove missing rows
df.dropna(inplace=True)

# remove duplicates
df.drop_duplicates(inplace=True)
```

We then clean price columns, discount percentages, clean rating and rating_count, and then verify by running the "df.info()" command.

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'].astype(str).str.replace('[‚Çπ,]','', regex=True), errors='coerce') # Convert 'discounted_price' to numeric, coercing errors to NaN
```

```python
df['actual_price'] = pd.to_numeric(df['actual_price'].astype(str).str.replace('[‚Çπ,]','', regex=True), errors='coerce') # Convert 'actual_price' to numeric, coercing errors to NaN 
```

```python
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].astype(str).str.replace('[%,]', '', regex=True), errors='coerce')
```

```python
df['rating']=pd.to_numeric(df['rating'], errors='coerce') # Convert 'rating' to numeric, coercing errors to NaN
```

```python
df['rating_count']=pd.to_numeric(df['rating_count'].astype(str).str.replace('[,]', '', regex=True), errors='coerce') # Convert 'rating_count' to numeric, coercing errors to NaN
```

```python
df.info()
```

## üìä VISUALIZATION 1 ‚Äî Category Dominance

We use bar chart to explore the number of products per category.
This is done to determine whether any category is dominating in sales, which would subsequently alter the insights as being biased towards that category.
This analysis also offers insights into whether our dataset is heavily baised towards a single category, or if we should focus our future analysis on that category only.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.countplot(
    y='main_category',
    data=df,
    order=df['main_category'].value_counts().index
)

plt.title("Number of Products per Category")
plt.xlabel("Count")
plt.ylabel("Category")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    y='sub_category',
    data=df,
    order=df['sub_category'].value_counts().index
)
plt.title("Number of Products per Sub-Category")
plt.xlabel("Count")
plt.ylabel("Sub-Category")
plt.show()
```

The above visualization gives us certain insights, such as the dominance of a few categories and the huge drop-off thereafter. This would suggest the demand as well as the pricing for those products, and their discounts drastically differ from the niche products, having much smaller markets.

## üìä VISUALIZATION 2 ‚Äî Price Distribution

Since histograms show the distribution of shape, we consider this graph to analyse how the prices of our products are distributed.
This shows us whether some products are super expensive and others super cheap, or not.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['actual_price'], bins=40)
plt.title("Distribution of Actual Prices")
plt.xlabel("Actual Price")
plt.ylabel("Frequency")
plt.show()
```

## üìä VISUALIZATION 3 ‚Äî Discount vs Rating

This is an important finding. This tells us whether giving more discounts on products would result in them getting better ratings. Scatterplots are ideal since they primarily tell relationships between entities or categories, in our case.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x='discount_percentage', y='rating', data=df)
plt.title("Discount Percentage vs Rating")
plt.xlabel("Discount Percentage")
plt.ylabel("Rating")
plt.show()
```

## üìä VISUALIZATION 4 ‚Äî Popularity vs Rating

This is another scatterplot chart, establishing relationship between ratings and the number of times a product is rated.
The main question to answer here is if a product reviewed many times would result in it getting higher ratings.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='rating_count', y='rating', data=df)
plt.title("Rating Count vs Rating")
plt.xlabel("Rating Count")
plt.ylabel("Rating")
plt.show()
```

Popular products would more often than not become stable near higher ratings, which doesn't seem to be he case here. Instead, less popular products are the ones which become more stable near average prices.

## üìä VISUALIZATION 5 ‚Äî Correlation Heatmap

Another important data analysis chart is the correlation chart, which signifies relationships between our categories simulatenously.

```python
plt.figure(figsize=(8,6))

corr = df[['actual_price','discounted_price','discount_percentage','rating','rating_count']].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```
